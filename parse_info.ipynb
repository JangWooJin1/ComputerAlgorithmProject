```python
import pandas as pd
import numpy as np

for i in range(11):
    df = pd.read_excel(f'data/raw/list{i+1}.xlsx', engine='openpyxl')
    df.head()
    my_data = np.array(df.iloc[:, [6, 2]])
    arr = []

    for data, target in my_data:
        if type(data) != float:
            data: str
            w: str  # 가중치
            val = data.split(',')
            for v in val:
                dow = v[0]
                dow_int = 0
                date = v[1:].split('-')

                if dow == '월':
                    dow_int = 0
                elif dow == '화':
                    dow_int = 1
                elif dow == '수':
                    dow_int = 2
                elif dow == '목':
                    dow_int = 3
                elif dow == '금':
                    dow_int = 4

                if target == '일교':
                    w = 1
                elif target == '공교':
                    w = 2
                elif target == '학기':
                    w = 3
                elif target == '전공':
                    w = 4
                elif target == '전필':
                    w = 5
                else:  # 에러 처리용도
                    w = 0
                arr.append((dow_int, float(date[0]), float(date[1]), w))
    arr.sort()

    with open(f'data/data/data{i+1}.txt','w') as f :
        for data in arr:
            f.write(f'{data[0]},{data[1]},{data[2]},{data[3]}\n')
```
